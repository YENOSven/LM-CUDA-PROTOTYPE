```python


device = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", device)

N = 4096  # try 2048 if you run out of VRAM

# CPU tensors
a_cpu = torch.rand(N, N)
b_cpu = torch.rand(N, N)

# GPU tensors
if device == "cuda":
    a_gpu = a_cpu.to("cuda")
    b_gpu = b_cpu.to("cuda")
    torch.cuda.synchronize()

# warmup (important)
_ = a_cpu @ b_cpu
if device == "cuda":
    _ = a_gpu @ b_gpu
    torch.cuda.synchronize()

# CPU timing
t0 = time.time()
c = a_cpu @ b_cpu
t1 = time.time()
print("PyTorch CPU:", t1 - t0)

# GPU timing
if device == "cuda":
    t0 = time.time()
    c = a_gpu @ b_gpu
    torch.cuda.synchronize()
    t1 = time.time()
    print("PyTorch GPU:", t1 - t0)
```
